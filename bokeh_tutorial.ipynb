Importing Libraries

```
import numpy as np
from bokeh.plotting import figure,show
from bokeh.io import output_notebook
```

Writing the Code

```
#preparing data
x = [1,3,4,5,6,9]
y = [7,9,8,11,12,14]

#output into this cell 
output_notebook()

#creating a new plot with title and axis labels
p = figure(title="First Bokeh Chart", x_axis_label="Day", y_axis_label="Temp")

#add a line renderer with legend and line thickness
p.line(x,y, legend="Temp", line_width=2)

#show the results
show(p)
```
